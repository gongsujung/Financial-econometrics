```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/계량/통합 문서3.csv')
df.columns = ['date', 'BM_month_rate', 'SDI_price', 'SDI_rate', 'LF_price',
              'LF_rate', 'SK_price', 'SK_rate', 'POSCO_price', 'POSCO_rate']
df.head()
```

```python
import matplotlib.pyplot as plt
from mlxtend.plotting import scatterplotmatrix
cols = ['BM_month_rate', 'SDI_price', 'SDI_rate', 'LF_price',
              'LF_rate', 'SK_price', 'SK_rate', 'POSCO_price', 'POSCO_rate']
scatterplotmatrix(df[cols].values, figsize=(18,15), names=cols, alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
from mlxtend.plotting import heatmap
import numpy as np
cols = ['BM_month_rate', 'SDI_price', 'SDI_rate', 'LF_price',
              'LF_rate', 'SK_price', 'SK_rate', 'POSCO_price', 'POSCO_rate']
cm = np.corrcoef(df[cols].values.T)
hm = heatmap(cm, row_names=cols, column_names=cols)
plt.show()
```
